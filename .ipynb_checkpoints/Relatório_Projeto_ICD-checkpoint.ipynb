# **Introdução**

A indústria de videogames é uma das mais lucrativas e de rápido crescimento no setor de entretenimento, atraindo milhões de jogadores em todo o mundo. Com o aumento das vendas e o surgimento de novas plataformas, entender os fatores que influenciam o sucesso dos jogos se tornou essencial para desenvolvedores, publicadores e investidores. O objetivo desta análise é examinar dados históricos de vendas de videogames, suas classificações e outras variáveis relacionadas para identificar padrões e tendências que possam fornecer insights sobre os fatores que impactam o desempenho de vendas e a popularidade dos jogos.

A motivação por trás desta análise é explorar como elementos como gênero, plataforma, classificação etária e localização geográfica influenciam as vendas, ajudando a responder perguntas como: Como as vendas de videogames variam por região (América do Norte, Europa, Japão, etc.)? Existe uma correlação entre a classificação de um jogo (ESRB Rating) e suas vendas? Jogos com avaliações mais altas (User Score, Critic Score) tendem a vender mais?

# **Dados Usados**


Para a análise, utilizamos o dataset "Video Game Sales with Ratings" do Kaggle, que contém informações detalhadas sobre jogos lançados desde a década de 1980 até 2016. Este dataset reúne mais de 16.000 registros de vendas globais de videogames, além de incluir dados sobre a plataforma, gênero, publicador, e a classificação de cada jogo, com base no sistema ESRB (Entertainment Software Rating Board).

# **Pré-processamento e Exploração Inicial dos Dados**

A primeira etapa da análise envolveu a vistoria do dataset para verificar sua estrutura e identificar possíveis erros. Na qual percebemos que haviam diversos valores que não haviam sido carregados corretamente, e constavam como NaN(Not a Number). Com isso decidimos excluir do dataset, todas as linhas que possuissem esse erro. Como mostrado a seguir:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('video_games.csv')


df.head()
```

```python
#Limpando o dataset de linhas que possuiam o valor "NaN"
#Possuia em torno de 9.894 linhas com esse valor 
df_limpo = df.dropna()
print(df_limpo)
```

Após isso, nós traduzimos o nome das colunas que originalmente estavam em inglês, para o português.

A partir da exclusão das linha com o NaN, foram excluídos do dataset 9.894 linhas. Sobrando 6.825 linhas de dados. E em cima dos dados que haviam sobrado, nós decidimos diminuir o scopo, e trabalhar com os dados de apenas os 100 jogos que possuiam as maiores quantidades de vendas globais.

```python
# Renomear as colunas do DataFrame para português
df_limpo = df_limpo.rename(columns={
    'Rank': 'Rank',  # Se preferir, pode manter "Rank" ou traduzir como "Classificação"
    'Name': 'Nome',
    'Platform': 'Plataforma',
    'Year': 'Ano',
    'Genre': 'Gênero',
    'Publisher': 'Publicadora',
    'NA_Sales': 'Vendas_NA',
    'EU_Sales': 'Vendas_EU',
    'JP_Sales': 'Vendas_JP',
    'Other_Sales': 'Vendas_Outros',
    'Global_Sales': 'Vendas_Globais',
    'Critic_Score': 'Nota_Crítica',
    'Critic_Count': 'Contagem_Crítica',
    'User_Score': 'Nota_Usuário',
    'User_Count': 'Contagem_Usuário',
    'Developer': 'Desenvolvedora',
    'Rating': 'Classificação'
})

# Exibir para verificar as colunas renomeadas
print(df_limpo.columns)
```

```python
# Ordenar o DataFrame df_limpo pelo número de vendas globais em ordem decrescente
df_sorted = df_limpo.sort_values(by='Vendas_Globais', ascending=False)

# Selecionar os top 100 jogos com maior venda global
top_100_jogos = df_sorted.head(100)

# Mostrar os 100 jogos com maior venda global
print(top_100_jogos)
```

Por fim, tomamos mais uma decisão de remoção de dados, removendo as colunas de números de votantes da nota da crítica, e a de usuário, não achavamos que seria relevante trabalhar com essas tabelas para o nosso projeto.

```python
# Removendo as colunas de números de votantes da nota da crítica, e a de usuário
top_100_jogos = top_100_jogos.drop(columns=['Contagem_Crítica', 'Contagem_Usuário'], errors='ignore')

# Verificar as colunas após a remoção
print("Colunas após remoção:", top_100_jogos.columns)
```

E daí partimos para as medidas de centralidade baseado no número de vendas das regiões e globalmente:

```python
colunas_numericas = ['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros', 'Vendas_Globais']


medidas = {}


for column in colunas_numericas:
    medidas[column] = {
        'Média': top_100_jogos[column].mean(),
        'Mediana': top_100_jogos[column].median(),
        'Desvio Padrão': top_100_jogos[column].std(),
        'Variância': top_100_jogos[column].var(),
        'Mínimo': top_100_jogos[column].min(),
        'Máximo': top_100_jogos[column].max(),
        'Primeiro Quartil (25%)': top_100_jogos[column].quantile(0.25),
        'Terceiro Quartil (75%)': top_100_jogos[column].quantile(0.75),
    }


medidas_df = pd.DataFrame(medidas).T
print(medidas_df)
```

Agora vamos fazer a análise dos boxplots: 

```python
import seaborn as sns
import matplotlib.pyplot as plt


numerical_columns = ['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros', 'Vendas_Globais']


plt.figure(figsize=(14, 8))


for i, column in enumerate(numerical_columns, start=1):
    plt.subplot(2, 3, i)  # Ajuste o layout conforme necessário
    sns.boxplot(x=top_100_jogos[column])
    plt.title(column)
    plt.xlabel('Valor')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

Vendas nas diferentes regiões: Cada região apresenta uma variação distinta nas vendas de videogames. Regiões como América do Norte e Europa geralmente possuem vendas mais altas, o que pode ser confirmado pela mediana elevada.

Outliers: É comum que alguns jogos se destaquem nas vendas, especialmente títulos extremamente populares como franquias de renome. Esses jogos aparecem como pontos fora dos bigodes nos boxplots, mostrando que há uma concentração de vendas em torno de certos valores, mas alguns jogos vendem significativamente mais que a maioria.

Comparação entre regiões: Os gráficos também permitirão comparar como o sucesso de um jogo varia de uma região para outra. Por exemplo, jogos de esportes podem ter vendas muito maiores em regiões específicas (como América do Norte ou Europa), enquanto jogos de RPG podem ser mais populares no Japão.

Uma analíse mais focada nas vendas globais, percebesse que a maior parte dos jogos tem vendas globais concentradas em torno da faixa de 10 a 20 milhões de unidades, como mostrado pela caixa do boxplot. Isso sugere que esses valores são típicos para jogos populares em termos de vendas globais. Os pontos fora do boxplot indicam jogos que venderam significativamente mais do que a maioria, o que é comum em casos de grandes franquias ou jogos extremamente populares. ## **1 - Como as vendas de videogames variam por região (América do Norte, Europa, Japão, etc.)?**

Para responder esta pergunta, decidimos usar o clássico gráfico de barras, que seria de fácil visualização e entendimento.

```python
vendas_por_regiao = top_100_jogos[['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros']].sum().reset_index()
vendas_por_regiao.columns = ['Região', 'Vendas (milhões)']

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='Região', y='Vendas (milhões)', data=vendas_por_regiao, palette='viridis')
plt.title('Vendas Totais de Videogames por Região (Top 100 Jogos)')
plt.xlabel('Região')
plt.ylabel('Vendas Totais (milhões)')
plt.grid(axis='y')
plt.show()
```

América do Norte (NA): As vendas nesta região são as mais altas, com cerca de 500 milhões de unidades vendidas. Essa região domina o mercado, representando uma grande parcela das vendas totais de videogames.
  Europa (EU): A segunda maior região em termos de vendas, com aproximadamente 400 milhões de unidades. Embora seja significativamente menor do que a América do Norte, ainda representa uma grande fatia do mercado.    Japão (JP): As vendas no Japão somam cerca de 100 milhões de unidades. Comparado com a América do Norte e a Europa, o mercado japonês é menor, mas ainda relevant
    Outras regiões: Essa categoria inclui regiões fora da América do Norte, Europa e Japão. As vendas totalizam também em torno de 100 milhões, similar ao Japão, o que sugere que outras regiões do mundo têm um impacto signific no número total de vendas globais.lusão:

A América do Norte é o maior mercado para videogames, seguida pela Europa. O Japão e outras regiões representam mercados menores, mas com vendas significativas. O gráfico evidencia uma clara concentração de vendas nas regiões ocidentais (NA e EU), enquanto o Japão e outras regiões têm um impacto relativamente menor.

## **2 - Existe uma correlação entre a classificação de um jogo (ESRB Rating) e suas vendas?**

```python
rating_mapping = {
    'E': 1,         # Livre para todas as idades
    'E10+': 2,     # Maiores de 10 anos
    'T': 3,         # Maiores de 13 anos
    'M': 4,         # Maiores de 17 anos
    'AO': 5,        # Maiores de 18 anos
    'RP': 6,        # Classificação pendente
}


df_limpo['Rating_Num'] = df_limpo['Classificação'].map(rating_mapping)


plt.figure(figsize=(12, 6))
sns.scatterplot(x='Rating_Num', y='Vendas_Globais', data=df_limpo)
plt.title('Dispersão: Classificação (ESRB Rating) vs. Vendas Globais')
plt.xlabel('Classificação (ESRB - numérica)')
plt.ylabel('Vendas Globais (milhões)')
plt.xticks(ticks=list(rating_mapping.values()), labels=list(rating_mapping.keys()))
plt.grid(True)
plt.show()
```


A maior concentração de jogos vendidos se encontra nas categorias:
E (Livre para todas as idades): Apresenta as vendas mais altas, com vários jogos acima de 20 milhões de unidades e pelo menos um jogo com vendas próximas de 80 milhões.
M (Maiores de 17 anos): Apesar de também ter muitos jogos, as vendas estão concentradas entre 0 e 20 milhões.

Classificações E10+ (Maiores de 10 anos), T (Maiores de 13 anos), AO (Maiores de 18 anos) e RP (Classificação pendente):
E10+ e T: As vendas são consideravelmente mais baixas, com a maioria dos jogos abaixo de 10 milhões de unidades.
AO (Maiores de 18 anos) e RP (Classificação pendente): Pouca representatividade, com alguns poucos pontos indicando baixas vendas. A categoria AO, por ser mais restrita, tem vendas especialmente limitadas.

Padrões Observados:
        Jogos classificados como "E" (Livre para todas as idades) tendem a alcançar maiores vendas. Isso porque são acessíveis para todas as idades, ampliando o público-alvo.
        Jogos M (Maiores de 17 anos), embora populares, parecem ter um teto mais baixo em termos de vendas máximas.
        Classificações mais restritivas (AO) ou incompletas (RP) têm pouca ou nenhuma influência no mercado de grandes vendas.

O gráfico sugere que jogos voltados para todos os públicos (classificação E) possuem maior sucesso comercial, enquanto jogos mais violentos ou maduros (Mature) têm vendas mais limitadas, apesar de também serem numerosos. Já os jogos AO e RP parecem ter baixa relevância comercial, refletindo as barreiras que essas classificações impõem ao mercado.

# Análise

Também criamos dois gráficos adicionais para enriquecer nossa análise: um gráfico de barras e um gráfico de pizza.

O gráfico de barras representa a quantidade de vendas em milhões, agrupadas por plataforma. Ele nos permite visualizar como cada plataforma contribui para as vendas totais dentro do top 100 jogos. Já o gráfico de pizza mostra a distribuição percentual de jogos por plataforma, ou seja, a proporção de jogos do top 100 que pertencem a cada plataforma.

Algo curioso pôde ser observado: no gráfico de pizza, percebemos que a plataforma com o maior número de jogos presentes no top 100 é o Xbox. No entanto, quando analisamos o gráfico de barras, fica claro que o Nintendo Wii teve um desempenho muito superior em termos de vendas totais. Isso indica que, embora o Nintendo Wii tenha menos jogos no top 100, os títulos que figuram nessa lista apresentam números de vendas consideravelmente maiores, o que reflete seu impacto significativo no mercado.

```python
plataforma_counts = top_100_jogos['Plataforma'].value_counts()


plt.figure(figsize=(10, 8))
plt.pie(plataforma_counts, labels=plataforma_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição das Plataformas (Top 100 Jogos)')
plt.axis('equal')  
plt.show()
```

```python
vendas_por_plataforma = top_100_jogos.groupby('Plataforma')['Vendas_Globais'].sum().sort_values(ascending=False)


plt.figure(figsize=(12, 6))
vendas_por_plataforma.plot(kind='bar', color='skyblue')
plt.title('Vendas Totais por Plataforma (Top 100 Jogos)')
plt.xlabel('Plataforma')
plt.ylabel('Vendas Globais (milhões)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
