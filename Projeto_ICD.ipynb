```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('video_games.csv')


df.head()
```

```python
df
```

```python
#Limpando o dataset de linhas que possuiam o valor "NaN"
#Possuia em torno de 9.894 linhas com esse valor 
df_limpo = df.dropna()
print(df_limpo)
```

```python
# Renomear as colunas do DataFrame para português
df_limpo = df_limpo.rename(columns={
    'Rank': 'Rank',  # Se preferir, pode manter "Rank" ou traduzir como "Classificação"
    'Name': 'Nome',
    'Platform': 'Plataforma',
    'Year': 'Ano',
    'Genre': 'Gênero',
    'Publisher': 'Publicadora',
    'NA_Sales': 'Vendas_NA',
    'EU_Sales': 'Vendas_EU',
    'JP_Sales': 'Vendas_JP',
    'Other_Sales': 'Vendas_Outros',
    'Global_Sales': 'Vendas_Globais',
    'Critic_Score': 'Nota_Crítica',
    'Critic_Count': 'Contagem_Crítica',
    'User_Score': 'Nota_Usuário',
    'User_Count': 'Contagem_Usuário',
    'Developer': 'Desenvolvedora',
    'Rating': 'Classificação'
})

# Exibir para verificar as colunas renomeadas
print(df_limpo.columns)
```

```python
# Ordenar o DataFrame df_limpo pelo número de vendas globais em ordem decrescente
df_sorted = df_limpo.sort_values(by='Vendas_Globais', ascending=False)

# Selecionar os top 100 jogos com maior venda global
top_100_jogos = df_sorted.head(100)

# Mostrar os 100 jogos com maior venda global
print(top_100_jogos)
```

```python
top_100_jogos
```

```python
# Removendo as colunas de números de votantes da nota da crítica, e a de usuário, não achavamos que seria necessário trabalhar com essas tabelas
top_100_jogos = top_100_jogos.drop(columns=['Contagem_Crítica', 'Contagem_Usuário'], errors='ignore')

# Verificar as colunas após a remoção
print("Colunas após remoção:", top_100_jogos.columns)
```

```python
top_100_jogos
```

```python

colunas_numericas = ['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros', 'Vendas_Globais']


medidas = {}


for column in colunas_numericas:
    medidas[column] = {
        'Média': top_100_jogos[column].mean(),
        'Mediana': top_100_jogos[column].median(),
        'Desvio Padrão': top_100_jogos[column].std(),
        'Variância': top_100_jogos[column].var(),
        'Mínimo': top_100_jogos[column].min(),
        'Máximo': top_100_jogos[column].max(),
        'Primeiro Quartil (25%)': top_100_jogos[column].quantile(0.25),
        'Terceiro Quartil (75%)': top_100_jogos[column].quantile(0.75),
    }


medidas_df = pd.DataFrame(medidas).T
print(medidas_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


numerical_columns = ['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros', 'Vendas_Globais']


plt.figure(figsize=(14, 8))


for i, column in enumerate(numerical_columns, start=1):
    plt.subplot(2, 3, i)  # Ajuste o layout conforme necessário
    sns.boxplot(x=top_100_jogos[column])
    plt.title(column)
    plt.xlabel('Valor')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

Fazendo um grafico de dispersão em relação ao número de vendas globais, com a classificação etária do jogo

```python

rating_mapping = {
    'E': 1,         # Livre para todas as idades
    'E10+': 2,     # Maiores de 10 anos
    'T': 3,         # Maiores de 13 anos
    'M': 4,         # Maiores de 17 anos
    'AO': 5,        # Maiores de 18 anos
    'RP': 6,        # Classificação pendente
}


df_limpo['Rating_Num'] = df_limpo['Classificação'].map(rating_mapping)


plt.figure(figsize=(12, 6))
sns.scatterplot(x='Rating_Num', y='Vendas_Globais', data=df_limpo)
plt.title('Dispersão: Classificação (ESRB Rating) vs. Vendas Globais')
plt.xlabel('Classificação (ESRB - numérica)')
plt.ylabel('Vendas Globais (milhões)')
plt.xticks(ticks=list(rating_mapping.values()), labels=list(rating_mapping.keys()))
plt.grid(True)
plt.show()
```

```python
#CRIANDO UM GRAFICO DE PIZZA QUE MOSTRA A DISTRIBUIÇAO DA PORCENTAGEM EM RELAÇÃO AS PLATAFORMAS DOS TOP 100 JOGOS
```

```python
# Contar a frequência de cada plataform
plataforma_counts = top_100_jogos['Plataforma'].value_counts()


plt.figure(figsize=(10, 8))
plt.pie(plataforma_counts, labels=plataforma_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição das Plataformas (Top 100 Jogos)')
plt.axis('equal')  
plt.show()
```

```python
#GRÁFICO DE BARRA PARA MOSTRAR AS VENDAS TOTAIS POR PLATAFORMA
```

```python
# Agrupar por plataforma e somar as vendas globais
vendas_por_plataforma = top_100_jogos.groupby('Plataforma')['Vendas_Globais'].sum().sort_values(ascending=False)


plt.figure(figsize=(12, 6))
vendas_por_plataforma.plot(kind='bar', color='skyblue')
plt.title('Vendas Totais por Plataforma (Top 100 Jogos)')
plt.xlabel('Plataforma')
plt.ylabel('Vendas Globais (milhões)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

GRÁFICO DE BARRA PARA MOSTRAR A VENDA TOTAIS POR REGIÃO

```python
vendas_por_regiao = top_100_jogos[['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros']].sum().reset_index()
vendas_por_regiao.columns = ['Região', 'Vendas (milhões)']

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='Região', y='Vendas (milhões)', data=vendas_por_regiao, palette='viridis')
plt.title('Vendas Totais de Videogames por Região (Top 100 Jogos)')
plt.xlabel('Região')
plt.ylabel('Vendas Totais (milhões)')
plt.grid(axis='y')
plt.show()
```

# K-means

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


vendas_cols = ['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros', 'Vendas_Globais']

# Separar os jogos com vendas globais > 20 milhões
cluster_maiores = top_100_jogos[top_100_jogos['Vendas_Globais'] > 20]
cluster_menores = top_100_jogos[top_100_jogos['Vendas_Globais'] <= 20]

# Padronizar os dados para jogos com vendas globais <= 20 milhões
scaler = StandardScaler()
scaled_data_menores = scaler.fit_transform(cluster_menores[vendas_cols])


kmeans = KMeans(n_clusters=1, random_state=42)
cluster_menores['cluster'] = kmeans.fit_predict(scaled_data_menores)


cluster_maiores['cluster'] = 1


top_100_jogos_clusterizado = pd.concat([cluster_menores, cluster_maiores])


plt.figure(figsize=(8, 6))
plt.scatter(top_100_jogos_clusterizado['Vendas_Globais'], top_100_jogos_clusterizado['Vendas_NA'], c=top_100_jogos_clusterizado['cluster'], cmap='viridis')

plt.xlabel('Vendas Globais (em milhões)')
plt.ylabel('Vendas América do Norte (em milhões)')
plt.title('Clusters de Vendas Regionais em comparação com Vendas Globais')
plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(top_100_jogos_clusterizado['Vendas_Globais'], top_100_jogos_clusterizado['Vendas_EU'], c=top_100_jogos_clusterizado['cluster'], cmap='viridis')

plt.xlabel('Vendas Globais (em milhões)')
plt.ylabel('Vendas Europa (em milhões)')
plt.title('Clusters de Vendas Regionais em comparação com Vendas Globais ')
plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(top_100_jogos_clusterizado['Vendas_Globais'], top_100_jogos_clusterizado['Vendas_JP'], c=top_100_jogos_clusterizado['cluster'], cmap='viridis')

plt.xlabel('Vendas Globais (em milhões)')
plt.ylabel('Vendas Japão (em milhões)')
plt.title('Clusters de Vendas Regionais em comparação com Vendas Globais ')
plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(top_100_jogos_clusterizado['Vendas_Globais'], top_100_jogos_clusterizado['Vendas_Outros'], c=top_100_jogos_clusterizado['cluster'], cmap='viridis')

plt.xlabel('Vendas Globais (em milhões)')
plt.ylabel('Vendas Outras Regiões (em milhões)')
plt.title('Clusters de Vendas Regionais em comparação com Vendas Globais ')
plt.show()
```

O gráfico mostra que as vendas no Japão estão concentradas majoritariamente abaixo de 2 milhões, com poucos jogos superando 4 milhões em vendas. Isso indica que, para a maioria dos jogos, o mercado japonês tem um impacto menor em comparação às vendas globais. E no Japão, novamente se confirma o fato de ter um impacto menor nas vendas globais, uma vez que muitos jogos com vendas globais expressivas não possuem um grande número de vendas no Japão.

Na Europa, as vendas regionais também são majoritariamente abaixo de 5 milhões, com algumas exceções de jogos que venderam até 10 milhões ou mais. As vendas na Europa parecem ser mais espalhadas em relação às vendas globais.

As vendas na região da América do Norte parecem estar mais concentradas, com a maioria dos jogos vendendo menos de 10 milhões na América do Norte, mas há jogos que se destacam com mais de 20 milhões em vendas.

E as demais regiões, também não possuem grande influência nas vendas globais, com a maioria dos jogos com valores de vendas menores que 4 milhões.

Uma pergunta que pode ser respondida foi "Existem jogos que têm sucesso em uma região específica, mas não globalmente?". No Japão fica bem claro isso, é o que mais possui jogos com um grande número de vendas local, mas não necessariamente um grande número de vendas no quesito global.
